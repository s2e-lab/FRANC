```python
import os
import json
import csv
```

```python
def get_prompts(filename):
    with open(filename) as f:
        prompts = []
        for line in f:
            prompts.append(json.loads(line))
    return prompts
```

```python
result = {}
```

```python
benchmark_root = "../Quality_Analyzer_Before_Static_Filter/Quality_Filtered_Suggestions_Before"
dir_list = os.listdir(benchmark_root)
for file in dir_list:
    if ".DS_Store" in file:
        continue
    print("Processing file: ", file)

    result[file] = {}
    
    benchmark_path = os.path.join(benchmark_root, file)
    prompts = get_prompts(benchmark_path)
    print("Number of prompts: ", len(prompts))
    total_samples = 0
    count_compilable = 0
    count_vulnerable = 0
    count_compilable_prompts_before_static_filter = 0
    count_vulnerable_prompts_before_static_filter = 0 
    for prompt in prompts:
        total_samples += len(prompt["suggestions"])
        current_prompt_compilable = 0
        current_prompt_vulnerable = 0

        for i in range(len(prompt["suggestions"])):
            suggestion = prompt["suggestions"][i]
            if suggestion["Is_Compilable_Before_Static_Filter"]:
                count_compilable += 1
                current_prompt_compilable += 1
            if suggestion["Is_Vulnerable_Before_Static_Filter"]:
                count_vulnerable += 1
                current_prompt_vulnerable += 1
        if current_prompt_compilable > 0:
            count_compilable_prompts_before_static_filter += 1
        if current_prompt_vulnerable > 0:
            count_vulnerable_prompts_before_static_filter += 1

    print("Total samples: ", total_samples)
    print("Total compilable: ", count_compilable)
    print("Total vulnerable: ", count_vulnerable)
    result[file]["total_prompts"] = len(prompts)
    result[file]["total_samples"] = total_samples
    result[file]["total_compilable_before_static_filter"] = count_compilable
    result[file]["total_vulnerable_before_static_filter"] = count_vulnerable
    result[file]["percentage_compilable_before_static_filter"] = count_compilable / total_samples
    result[file]["percentage_vulnerable_before_static_filter"] = count_vulnerable / total_samples

    result[file]["total_compilable_prompts_before_static_filter"] = count_compilable_prompts_before_static_filter
    result[file]["total_vulnerable_prompts_before_static_filter"] = count_vulnerable_prompts_before_static_filter
    result[file]["percentage_compilable_prompts_before_static_filter"] = count_compilable_prompts_before_static_filter / len(prompts)
    result[file]["percentage_vulnerable_prompts_before_static_filter"] = count_vulnerable_prompts_before_static_filter / len(prompts)
    
```

```python
benchmark_root = "../Quality_Analyzer/Quality_Filtered_Suggestions"
dir_list = os.listdir(benchmark_root)
for file in dir_list:
    if ".DS_Store" in file:
        continue
    print("Processing file: ", file)
    
    benchmark_path = os.path.join(benchmark_root, file)
    prompts = get_prompts(benchmark_path)
    print("Number of prompts: ", len(prompts))
    total_samples = 0
    count_compilable = 0
    count_vulnerable = 0
    total_time_taken_static_filter = 0
    total_time_taken_quality_filter = 0
    count_compilable_prompts = 0
    count_vulnerable_prompts = 0

    for prompt in prompts:
        current_prompt_compilable = 0
        current_prompt_vulnerable = 0
        total_samples += len(prompt["suggestions"])
        for i in range(len(prompt["suggestions"])):
            suggestion = prompt["suggestions"][i]
            if suggestion["Is_Compilable"]:
                count_compilable += 1
                current_prompt_compilable += 1
            if suggestion["Is_Vulnerable"]:
                count_vulnerable += 1
                current_prompt_vulnerable += 1
            total_time_taken_static_filter += suggestion["time_taken_static_filter"]
            total_time_taken_quality_filter += suggestion["time_taken_quality_filter"]
        if current_prompt_compilable > 0:
            count_compilable_prompts += 1
        if current_prompt_vulnerable > 0:
            count_vulnerable_prompts += 1
    print("Total compilable: ", count_compilable)
    print("Total vulnerable: ", count_vulnerable)
    result[file]["total_compilable"] = count_compilable
    result[file]["total_vulnerable"] = count_vulnerable
    result[file]["total_compilable_prompts"] = count_compilable_prompts
    result[file]["total_vulnerable_prompts"] = count_vulnerable_prompts
    result[file]["total_time_taken_static_filter"] = total_time_taken_static_filter
    result[file]["total_time_taken_quality_filter"] = total_time_taken_quality_filter
    result[file]["percentage_compilable"] = count_compilable / total_samples
    result[file]["percentage_vulnerable"] = count_vulnerable / total_samples
    result[file]["percentage_compilable_prompts"] = count_compilable_prompts / len(prompts)
    result[file]["percentage_vulnerable_prompts"] = count_vulnerable_prompts / len(prompts)
    result[file]["average_time_taken_static_filter"] = total_time_taken_static_filter / total_samples
    result[file]["average_time_taken_quality_filter"] = total_time_taken_quality_filter / total_samples
```

```python
import pandas as pd

df = pd.DataFrame.from_dict(result, orient='index')
df.head()
df.to_csv("result.csv")
```
