```python
import pandas as pd
```

```python
df = pd.read_csv("Prompt_Number_NDCG.csv")
df.head()
```

```python
old_top_df = df[df["Old_Top_1_Score"] == 3]
old_top_df.head()
```

```python
def get_dataset(x):
    if 'SOEval' in x:
       return "_".join(x.split('_')[0:1])
    else:
        return "_".join(x.split('_')[0:2])
```

```python
def get_model(x):
    if 'SOEval' in x:
        return x.split('_')[1]
    else:
        return x.split('_')[2]
```

```python
def get_language(x):
    dataset_type = "Java"
    if "python" in x.lower():
        dataset_type = "Python"
    return dataset_type
```

```python
df.insert(1, "Dataset", df["Benchmark"].apply(get_dataset))
df.insert(2, "Model", df["Benchmark"].apply(get_model))
df.insert(3, "Language", df["Benchmark"].apply(get_language))
```

```python
df.head()
df.to_csv("Score_data.csv", index=False)
```

```python
old_grouped_df = old_top_df.groupby(["Model","Language"]).count()
old_grouped_df.head()
```

```python
old_grouped_df.to_csv("old_grouped_old_top_1.csv")
```
