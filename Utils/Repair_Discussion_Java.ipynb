```python
import os
import json
import csv
```

```python
def get_prompts(filename):
    with open(filename) as f:
        prompts = []
        for line in f:
            prompts.append(json.loads(line))
    return prompts
```

```python
result = {}
fix_keys = {}
benchmark_root = "../Repair_Quality_Analyzer/Repair_Quality_Filtered_Suggestions_2/"
dir_list = os.listdir(benchmark_root)
for file in dir_list:
    if ".DS_Store" in file:
        continue
    print("Processing file: ", file)
    result[file] = {}
    prompts = get_prompts(benchmark_root + file)
    print("Number of prompts: ", len(prompts))
    if len(prompts) == 0:
        continue
    for prompt in prompts:
        to_be_fixed = []
        check_not_fixed = {}
        for line in prompt["repair_prompt"].split("\n"):
            if line.startswith("// Fix: "):
                # print(",".join(line.split(", ")[1:]))
                if "At line " in line:
                    fix_key = ",".join(line.split(", ")[1:])
                else:
                    fix_key = ",".join(line.split(": ")[1:])
                if fix_key not in fix_keys:
                    fix_keys[fix_key] = {
                        "total": 1,
                        "fixed_prompt": 0,
                        "fixed_suggestion": 0,
                    }
                else:
                    fix_keys[fix_key]["total"] += 1

                if fix_key not in check_not_fixed:
                    check_not_fixed[fix_key] = 10
                to_be_fixed.append(fix_key)
        if len(to_be_fixed) > 0:
            result[file][prompt["task_id"]] = to_be_fixed
        else:
            continue
        for issue_text in check_not_fixed:
            total = 0
            for i in range(len(prompt["suggestions"])):
                suggestion = prompt["suggestions"][i]
                if not suggestion["Is_Compilable"]:
                    check_not_fixed[issue_text] -= 1
                    continue
                found = False
                if "Analyzer_Result" not in suggestion or suggestion["Analyzer_Result"] is None:
                    continue
                if "BugCollection" not in suggestion["Analyzer_Result"] or suggestion["Analyzer_Result"]["BugCollection"] is None:
                    continue
                if "BugInstance" not in suggestion["Analyzer_Result"]["BugCollection"] or suggestion["Analyzer_Result"]["BugCollection"]["BugInstance"] is None:
                    continue
                if type(suggestion["Analyzer_Result"]["BugCollection"]["BugInstance"]) is list:
                        for j in range(len(suggestion["Analyzer_Result"]["BugCollection"]["BugInstance"])):
                            if issue_text in suggestion["Analyzer_Result"]["BugCollection"]["BugInstance"][j]["ShortMessage"]:
                                found = True
                else:
                    if issue_text in suggestion["Analyzer_Result"]["BugCollection"]["BugInstance"]["ShortMessage"]:
                        found = True

                if found:
                    total += 1
            check_not_fixed[issue_text] = check_not_fixed[issue_text] - total
                    
        for key in check_not_fixed:
            if check_not_fixed[key] >0:
                fix_keys[key]["fixed_prompt"] += 1
            fix_keys[key]["fixed_suggestion"] += check_not_fixed[key]

    if len(result[file]) == 0:
        del result[file]

print("Number of files: ", len(result))
print("Number of fix keys: ", fix_keys)
```

```python
for key in fix_keys:
    fix_keys[key]["Percentage_fixed_prompt"] = (fix_keys[key]["fixed_prompt"] / fix_keys[key]["total"])*100
```

```python
import pandas as pd

df = pd.DataFrame.from_dict(fix_keys)
df.to_csv("repair_discussion_result_java_2.csv")
```
